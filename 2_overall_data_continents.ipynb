#### 3. Barplots

```python
for c in data.select_dtypes(include=np.number):
  if c == 'Year':
    continue
  else:    
    fig = plt.figure(figsize=(10, 5))

    barplot = sns.barplot(
            data=data,
            x='WHO Region',
            y=c,
            dodge=False
        )
    
    barplot.tick_params(
            labelsize=10, 
            pad=5)
    
    sns.despine()

    plt.title(
            f'{c}',
            fontsize=15,
            fontweight='bold',
            pad = 20)

    plt.ylabel(
            'Count',
            fontsize=15
        )

    plt.xlabel(
            'Region according with World Health Organization',
            fontsize=15
        )

    plt.show()
    print('\n')
```
