```python
from module import data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set_theme(context='paper', palette='deep')
sns.set_style('ticks')
```

## Exploratory data analysis - Western Pacific

#### 1. Sum of cases, deaths, and fatality rate of cholera in Western Pacific countries

```python
data_western = data.loc[data['WHO Region']=='Western Pacific']

for c in data_western.select_dtypes(include=np.number):
  if c == 'Year':
    continue
  else:    
    fig = plt.figure(figsize=(5,10))

    barplot = sns.barplot(
            data=data_western,
            x=c,
            y='Country',
            dodge=False
        )
    
    barplot.tick_params(
            labelsize=15, 
            pad=5)
    
    sns.despine()

    plt.title(
            f'{c} - Western Pacific',
            fontsize=15,
            fontweight='bold',
            pad = 20)

    plt.ylabel(
            'Country',
            fontsize=15
        )

    plt.xlabel(
            c,
            fontsize=15
        )

    plt.show()
    print('\n')
```

#### 2. Distribution of cases, deaths, and fatality rate over the years

```python
for i in data_western.select_dtypes(include=np.number):
    
    if i != 'Year':
        fig = px.scatter(
            data_frame = data_western,
            x='Year',
            y=i,
            color='Country',
            width=800,
            title=i
        )

        fig.show()
```

#### 2. Number of cases and deaths per country per year

```python
for i in data_western['Country'].unique():

    fig, axes = plt.subplots(1, 2, figsize=(7,3))

    sns.lineplot(
        data=data_western.loc[data['Country']==i],
        x='Year',
        y='Number of reported cases of cholera',
        ax=axes[0])

    axes[0].set(
        title=f'{i} - Cases'
    )

    sns.lineplot(
        data=data_western.loc[data['Country']==i],
        x='Year',
        y='Number of reported deaths from cholera',
        ax=axes[1])
    
    axes[1].set(
        title=f'{i} - Deaths'
    )
    fig.tight_layout(pad=0.8)
    sns.despine()
    plt.show()
```

#### 3. Cases vs Deaths per country

```python
for i in data_western['Country'].unique():

    fig = px.bar(
    data_frame = data_western.loc[data['Country']==i], 
    x='Year', 
    y='Number of reported cases of cholera', 
    width=1000, 
    color='Number of reported deaths from cholera',
    color_discrete_sequence ='viridis',
    title=f'{i} - Cases vs Deaths')

    fig.show()
```

#### 3. Correlation heatmap

```python
sns.heatmap(
    data=data_western.corr(numeric_only=True),
    cmap='inferno',
    annot=True)

plt.show()
```

## Conclusions:

- The fatality rate appears to have decreased after the 90's, perhaps due to the oral vaccine that came out around this time. There was an slight increase in cases and deaths in the early 90's, maybe because of some outbreak.
- Similar to other WHO regions, there are countries lacking information on cholera, like Kiribati, Marshall Islands, North Korea, Samoa, and others.
- The number of cases and the number of deaths appear to have some correlation, but further statistical analysis are needed to confirm its significance.
